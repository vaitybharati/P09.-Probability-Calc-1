# Probability Calc-1

#### Find the probability that a normally distributed random variable has a mean of 60 and a standard deviation of 10 and we want to find the probability of x is less than 70. 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
dist=np.linspace(30,90,1000) # linespace(start point, stop point, no. of datapoints) # dist= distribution
plt.plot(dist,norm.pdf(dist,60,10)) # norm.pdf(distribution, mean, standard deviation)
plt.title("Standard normal")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from scipy import stats
```

```python
# For P(X<70)
stats.norm.cdf(70,60,10) # stats.norm.cdf(x,loc=mean,scale=SD)
```

```python
# For P(X>70)
1-stats.norm.cdf(70,60,10)
```

```python
# For P(X>80)
1-stats.norm.cdf(80,60,10)
```
